```python
# Dependencies and Setup
import pandas as pd

# File to Load (Remember to Change These)
file_to_load = "Resources/purchase_data.csv"

# Read Purchasing File and store into Pandas data frame
purchase_data = pd.read_csv(file_to_load)

purchase_data.head()
```

```python
#Check to see if there are any rows with missing data
purchase_data.count()
```

```python
print("Total Players")
print("______________")
purchase_data['SN'].nunique()
```

```python
unique_items = purchase_data['Item Name'].nunique()
average_price = purchase_data['Price'].mean()
number_of_purchases = purchase_data['Purchase ID'].count()
total_revenue = purchase_data['Price'].sum()

```

```python
summary_purchase_data = pd.DataFrame({"Number of Unique Items" : [unique_items],
                                      "Average Price" : [average_price],
                                      "Number of Purchases" : [number_of_purchases],
                                      "Total Revenue" : [total_revenue]
    
})

summary_purchase_data
```

```python
 screen_df = purchase_data.drop_duplicates(subset=['SN'])

Total_Count_Gender= screen_df['Gender'].value_counts()
Percentage_0f_Players =screen_df['Gender'].value_counts(normalize=True).mul(100).round(1).astype(str) + '%'
df = pd.DataFrame({'Total Count': Total_Count_Gender, 'Percentage of Players': Percentage_0f_Players})
    
print(df)
```

```python
purchase_gender = purchase_data.groupby("Gender")
Purchase_Count = purchase_gender['Purchase ID'].count()
average_purchase_gender = purchase_gender["Price"].mean() 
total_gender_purchase = purchase_gender['Price'].sum()
avg_per_total = total_gender_purchase / Total_Count_Gender

print(avg_per_total)
```

```python
sf = pd.DataFrame({'Purchase Count': Purchase_Count, 'Average Purchase Price' : average_purchase_gender, 'Total Purchase Value' : total_gender_purchase, 'Avg Total Purchase Per Person':avg_per_total})
                                      
    
sf
```

```python
bins = [0, 10, 14, 19, 24, 29, 34, 39, 40]
group_labels = ["< 10", "10-14", "15-19" , "20-24", "25-29", "30-34", "35-39", "40"]
pd.cut(screen_df["Age"], bins, labels=group_labels).head()
```

```python
screen_df["age_ranges"] = pd.cut(screen_df["Age"], bins, labels=group_labels)
age_group = screen_df.groupby("age_ranges")
total_Age_count = age_group["Age"].sum()
total_Age_percent=age_group["Age"].sum() * 100


Af = pd.DataFrame({'Total Count': total_Age_count, 'Percentage of Players' : total_Age_percent})

Af
                                      

```

```python
Purchase_Count_age = age_group['Purchase ID'].count()
average_purchase_age = age_group["Price"].mean() 
total_age_purchase=age_group['Price'].sum()
avg_per_total_age = total_age_purchase / total_Age_count

AGf = pd.DataFrame({'Purchase Count': Purchase_Count_age, 'Average Purchase Price' : average_purchase_age, 'Total Purchase Value' : total_age_purchase, 'Avg Total Purchase Per Person':avg_per_total_age})

AGf
                                      

```

```python
top_spenders = purchase_data.groupby("SN")
Total_Count_SN= top_spenders['SN'].value_counts()
Purchase_Count_SN = top_spenders['Purchase ID'].count()
average_purchase_SN = top_spenders["Price"].mean() 
total_SN_purchase = top_spenders['Price'].sum()
avg_per_total = total_SN_purchase / Total_Count_SN

SNF = pd.DataFrame({'Purchase Count': Purchase_Count_SN, 'Average Purchase Price' : average_purchase_SN, 'Total Purchase Value' : total_SN_purchase})

SNF
               
```
